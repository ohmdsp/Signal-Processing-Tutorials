# Introduction to Detection, Estimation and Classification Theory


In a typical detection and estimation problem, our goal is to decide when an event of interest occurs, and then deremine more information about that event. The former problem, that of decision making, is broadly termed _detection theory_. The latter problem, that of determining more information, is the subject of _estimation theory_.

* In the context of signal processing, when using detection theory we seek to determine the presence or absence of a signal using the measured data. Example: Is there a person in this video frame?, or is it a 1 or 0 from a BPSK waveform?
* When using estimation theory, we seek to estimate the values of a parameter(s) from measured data. These parameter values affect the distribution of the measured data. Example data could be: radar, sonar, speech, imagery, communications, biology, etc. Example paramters could be time-of-arival (TOA), speed and direction, frequency, ampitude and phase, etc. 

### Section 1: Detection Theory Basics

Consider the detection of a DC signal in white gaussian noise where $A = 1$ and $w[0] \sim \cal N(0,1)$.

* The hypothesis could be
\begin{align*}
{\cal H}_0 &: x[n] = w[n] \\
{\cal H}_1 &: x[n] = A + w[n]
\end{align*}               
  
  
* Consider the simple example of observing only one sample $x[0]$ from this process. This process can be modeled by two normally distributed pdfs (i.e., $\cal N(\mu ,\sigma^2)$ with the pdfs: $\cal N(0,1)$ or $\cal N(1,1)$. Our job is to determine which pdf best represents our sample.
* Our hypothesis could also be 
\begin{align*}
{\cal H}_0 &: \mu = 0,\\
{\cal H}_1 &: \mu = 1,
\end{align*}


Note the following definitions:
* Deciding ${\cal H}_1$ when ${\cal H}_0$ is true can be considered a _false alarm_. (Note: also called a _Type I Error_, or _false positive_)
* $P({\cal H}_1;{\cal H}_0)$ is refered to as the _probability of false alarm_, or $P_{fa}$.
* The other error, deciding ${\cal H}_0$ when ${\cal H}_1$ is true is called a _Type II Error_, or _false negative_. 
* The probability of this error can be written as $P({\cal H}_0;{\cal H}_1)$ and is sometimes refered to as a probablility of a miss $P_{miss}$.
* *Note*: It is not possible to reduce both $P({\cal H}_1;{\cal H}_0)$ and $P({\cal H}_0;{\cal H}_1)$ simultaneously. 

#### White Gaussian Process

Let's look at the _Gaussian_ probability density function (PDF) for a white gaussian process. This is called the normal distribution. It is very convenient to model your problem so that this pdf is represents the distribution of parameter values you are interested in. More on that later. 

$$ p(x) = \frac{1}{\sqrt{2\pi \sigma^{2}}}\exp\left(-\frac{(x -\mu)^2}{2\sigma^{2}}\right)  \qquad   -\infty < x < \infty $$

It is important to know the following:
* The width of the distribution is determined by the variance, or $\sigma^{2}$
* The standard deviation is the square root of the variance. 68% of the random parameter values will be within on standard deviation of the mean. 95% will be within 2 standard deviations.  

```python
# First, let's look at how to plot the pdf for gaussian white noise process (i.e., normal distribution) 

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = (12, 4)

# get x values for mean and std deviation and setup normal pdf formula
x = np.linspace(-5,5,5000)
mu = 0
sigma = 1
y = (1/np.sqrt(2*np.pi*np.power(sigma,2)))*(np.power(np.e,-(np.power((x-mu),2)/(2*np.power(sigma,2)))))

#plot the pdf
plt.plot(x,y)
```

```python
# Plot two normal distributions for two processes (H0 & H1) using scipy
import scipy.stats as ss

def plot_normal(x_range, mu=0, sigma=1, **kwargs):
    '''
    Plots the normal distribution function for a given x range
    If mu and sigma are not provided, standard normal is plotted
    '''
    x = x_range
    y = ss.norm.pdf(x, mu, sigma)
    plt.plot(x, y, **kwargs)


x = np.linspace(-5, 5, 5000)
plot_normal(x, 0, 1, color='blue', lw=2, ls='-', alpha=0.5)
plot_normal(x, 1, 1, color='red', lw=2, ls='-', alpha=0.5)
```

#### Probability of Detection

* $P({\cal H}_1;{\cal H}_1)$, or equivalently $1 - P({\cal H}_0;{\cal H}_1)$ is called the _probability of detection_, or $P_{D}$.
* Often the goal is to maximize $P_{d}$ subject to the contraint $P_{fa} = \alpha$.
* This setup is termed the _Neyman-Pearson_ method of signal detection. 

Let's return to our previous example and assume that you use a single sample x[0] for detection. An obvious method for deciding which process the sample came from would be to choose the distribution that corresponds with the x[0] value being over a certain threshold (example: $x[0] > 1$). This is essentially choosing the most "likely" one. 

The associated Gaussian pdfs, or likelihood functions, are:
\begin{align*}
{\cal H}_1 &: p(x[0];\mu=1) =  \frac{1}{\sqrt{2\pi}}\exp\left(-\frac{(x[n]-1)^2}{2}\right).\\
{\cal H}_0 &: p(x[0];\mu=0) =  \frac{1}{\sqrt{2\pi}}\exp\left(-\frac{(x[n])^2}{2}\right).
\end{align*}


Note - The idea is that you should select ${\cal H}_1$ if "$\mu = 1$" is more likely than "$\mu = 0$", which can be expressed mathematically as $p(x[0];\mu=1) > p(x[0];\mu=0)$. We can rearrange and this expression and write this as a ratio. Then, fill in the details of each pdf. 

\begin{align*}
\frac{p(x[0];\mu=1)}{p(x[0];\mu=0)} > 1 \\
\\
\frac{\frac{1}{\sqrt{2\pi}}\exp\left(-\frac{(x[n]-1)^2}{2}\right)}{\frac{1}{\sqrt{2\pi}}\exp\left(-\frac{(x[n])^2}{2}\right)} > 1 \\
\\
\exp\left(-\frac{(x[n]-1)^2 - x[n]^2}{2}\right) > 1
\\
\end{align*}

\begin{align*}
\\
(x[n]^2 - (x[n]-1)^2) > 0\\\\
2x[n] - 1 > 0\\\\
x[n] > \frac12.
\end{align*}


In other words, choose ${\cal H}_1$ if $x[0] > 0.5$ and ${\cal H}_0$ if
$x[0] < 0.5$.

The ratio used above is called the _likelihood ratio_, and comparison to a threshold $\gamma$ (example above $\gamma = 1$) is called _likelihood ratio test_ (LRT).

Of course, the threshold $\gamma$ may be chosen other than $\gamma = 1$. It might be that the detection problem is not symmetric and some errors are more costly than others. For example, when detecting a balistic missle, a missed detection is more costly than a false alarm. The tradeoff between misses and false alarms can be adjusted using the threshold of the LRT.

The below figure illustrates the probabilities of the two kinds of errors. The blue area corresponds to the probability of choosing ${\cal H}_1$ while ${\cal H}_0$ is true (false alarm). The red area is the probability of choosing  ${\cal H}_0$ while ${\cal H}_1$ is true (missed detection). Both probabilities can be calculated.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
norm.cdf(1.96)

def solve(m1,m2,std1,std2):
  a = 1/(2*std1**2) - 1/(2*std2**2)
  b = m2/(std2**2) - m1/(std1**2)
  c = m1**2 /(2*std1**2) - m2**2 / (2*std2**2) - np.log(std2/std1)
  return np.roots([a,b,c])

m1 = 0
std1 = 1.0
m2 = 1
std2 = 1.0

#Get point of intersect
result = solve(m1,m2,std1,std2)

#Get point on surface
x = np.linspace(-5,5,5000)
plot1=plt.plot(x,norm.pdf(x,m1,std1),color='blue')
plot2=plt.plot(x,norm.pdf(x,m2,std2),color='red')
plot3=plt.plot(result,norm.pdf(result,m1,std1),'o')

#Plots integrated area
r = result[0]
olap = plt.fill_between(x[x>r], 0, norm.pdf(x[x>r],m1,std1),alpha=0.3,color='blue')
olap = plt.fill_between(x[x<r], 0, norm.pdf(x[x<r],m2,std2),alpha=0.3,color='red')

# integrate
area = norm.cdf(r,m2,std2) + (1.-norm.cdf(r,m1,std1))
print("Area under curves ", area)

plt.show()
```

Examples: (note: Blue = False Alarms, Red = Missed Detects)

For the example given above, the probability of false alarm for the threshold $\gamma = 1.5$ is $$
P_{FA} = P(x[0]>\gamma ; \mu = 0) = \int_{1.5}^{\infty} \frac{1}{\sqrt{2\pi}}\exp\left(-\frac{(x[n])^2}{2}\right)\,dx[n]\approx 0.0668.
$$

Probability of missed detection is
$$
P_{M} = P(x[0]<\gamma ; \mu = 1) = \int_{-\infty}^{1.5} \frac{1}{\sqrt{2\pi}}\exp\left(-\frac{(x[n]-1)^2}{2}\right)\,dx[n] \approx 0.6915.
$$

An equivalent, but more useful term is the complement of $P_M$: probability of detection:

$$
P_{D} = 1 - P_M = \int_{1.5}^{\infty} \frac{1}{\sqrt{2\pi}}\exp\left(-\frac{(x[n]-1)^2}{2}\right)\,dx[n] \approx 0.3085.
$$


#### Choosing a Threshold
Since $P_{FA}$ and $P_D$ depend on each other, we would like to maximize
$P_D$ subject to given maximum allowed $P_{FA}$. Luckily the following theorem makes this easy.

$\textbf{Neyman-Pearson Theorem}$: For a fixed $P_{FA}$, the likelihood ratio test
maximizes $P_D$ with the decision rule
$$
L(x) = \frac{p(x; {\cal H}_1)}{p(x; {\cal H}_0)} > \gamma,
$$
with threshold $\gamma$ is the value for which
$$
\int_{x:L(x)>\gamma} p(x; {\cal H}_0)\, dx = P_{FA}.
$$

As an example, suppose we want to find the best detector for our
introductory example, and we can tolerate 10\% false alarms ($P_{FA} = 0.1$).
According to the theorem, the detection rule is:

$$
{\cal H}_1 \text{ if } \frac{p(x\mid \mu=1)}{p(x\mid \mu=0)} > \gamma
$$
The only thing to find out now is the threshold $\gamma$ such that
$$
\int_{\gamma}^{\infty} p(x\mid \mu=0)\, dx = 0.1.
$$

This can be done with Python function _isf_, which solves the inverse cumulative distribution function.
Note - The parameters _loc_ and _scale_ are the mean and standard
deviation of the Gaussian density, respectively.

```python
import scipy.stats as stats
# Compute threshold such that P_FA = 0.1
Thresh = stats.norm.isf(0.1, loc = 0, scale = 1.0)
print(Thresh)
```

#### Types of Receivers
When dealing with complex signals and noises, there are two types of receivers. The first kind is a coherent receiver, which assumes that both the amplitude and phase of the received signal are known. The output samples from a coherent receiver are complex valued. However, the detector following the matched filter in a coherent receiver normally uses only the real part of the signal. A coherent receiver is vulnerable to phase errors. 

For a noncoherent receiver, the received signal is modeled as a copy of the original signal with a random phase error. With a noncoherent received signal, the detection after the matched filter is normally based on the power or magnitude of the signal since you need both real and imaginary parts to completely define the signal.


#### Matched Filter 
When detection a known deterministic signal (i.e., no unkown parameters) in WGN the optimal NP detector is called the _matched filter_. I will show you the derivation next time. It is

$$ 
T(x) = \sum_{n-0}^{N-1} x[n]s[n]>\gamma 
$$

where $T(x)$ is our test statistic (a funtion of the data) and $s[n]$ is a replica of the signal we want to detect. Again the threshold is chosen to satisfy $P_{fa} = \alpha$ for a given performance $\alpha$. This is also reffered to as a _correlator_ or _matched correlator_ since we correlate the received data with a replica of the signal. 

#### Section 2: Estimation Theory Basics

Let's assume that a our mathematical model for a signal is 
$$
s[n] = A + Bn
$$ 

Assuming actual data $x[n]$ for $n=0,1,...,N-1$ is available, the parameters A and B could be found easily by using a least squares approach. It determines A and B by finding the best match between the data and the signal model in a least squares sense. Specifically, we would choose the A and B parameter values that minimize the squared error.

$$
J(A,B) = \sum_{n=0}^{N-1}(x[n]-(A+Bn))^2
$$

Since J is a quadratice function of A and B, the values that minimize J can be found by taking the partial dirivatives of J with respect to A and B and setting them equal to zero. 

Example: Assume that $B=0$ so that $J=\sum_{n=0}^{N-1}(x[n]-A)^2$. Find the estimate of $A$ (call it $\hat{A}$).

$$
\frac{dJ(A)}{dA} = -2\sum^{N-1}_{n=0}(x[n]-A) = 0
$$
Solving for the parameter estimate yields
$$
\hat{A} = \frac{1}{N}\sum^{N-1}_{n=0}x[n]
$$
which is the sample mean. This is a reasonable estimator for a DC signal embeded is zero mean noise. It is easy to show that if $x[n] = A + w[n]$, where $w[n]$ is WGN with variance $\sigma^{2}$, then 
$$
var(\frac{1}{N}\sum^{N-1}_{n=0}w[n]) = \frac{\sigma^2}{N}
$$
or in other words, the noise power is reduced by $N$ due to averaging.
